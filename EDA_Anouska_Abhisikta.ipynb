```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/Placement_Data_Full_Class.csv")
df.head()
```

```python
plt.figure(figsize=(10,5))
#Count of people in various secondary school boards
plt.subplot(1,3,1)
plt.title('Count_secondary_school_board')
sns.countplot(x = 'ssc_b', palette = 'Set3', data = df)
```

```python
plt.figure(figsize = (10,5))
sns.histplot(df['ssc_p'])
```

```python
#boxplot analysis of secondary school percentage
plt.figure(figsize = (10,5))
plt.title("Boxplot secondary school_percentage")
sns.boxplot(x=df["ssc_p"])
```

```python
plt.figure(figsize = (10, 8))
#plot between secondary school board and placement status
plt.subplot(2,2,1)
plt.title('Status_secondary_school_board')
sns.countplot(x = 'status', hue = 'ssc_b', palette = 'Set1', data = df)
```

```python
plt.figure(figsize = (10, 8))
#plot between gender and various secondary school boards
plt.subplot(2,2,3)
sns.countplot(x = 'gender', hue = 'ssc_b', palette = 'Set1', data = df)
```

```python
plt.figure(figsize = (10,5))
#box plot analysis between secondary school board percentage and placement status
plt.title('Status_high_school_board')
sns.boxplot(x="status", y="ssc_p", data=df)
```

```python
plt.figure(figsize = (10, 15))
plt.suptitle(" Boxplot Analysis Of Variable secondary_school_percentage",fontweight="bold")

plt.subplot(3,2,1)
sns.boxplot(x="gender", y="ssc_p", data=df)

plt.subplot(3,2,2)
sns.boxplot(x="ssc_b", y="ssc_p", data=df)

plt.subplot(3,2,3)
sns.boxplot(x="ssc_b", y="hsc_p", data=df)

plt.subplot(3,2,5)
sns.boxplot(x="degree_t", y="ssc_p", data=df)
```

```python
plt.figure(figsize = (10, 15))
plt.suptitle("Scatterplot Analysis Of Variable secondary school percentage",fontweight="bold", fontsize=20)

plt.subplot(2,2,1)
sns.scatterplot(data=df, x="hsc_p", y="ssc_p", palette = 'Set2', hue = 'status')

plt.subplot(2,2,2)
sns.scatterplot(data=df, x="ssc_p", y="degree_p", palette = 'Set2', hue = 'status')

plt.subplot(2,2,3)
sns.scatterplot(data=df, x="ssc_p", y="etest_p", palette = 'Set2', hue = 'status')

plt.subplot(2,2,4)
sns.scatterplot(data=df, x="ssc_p", y="mba_p", palette = 'Set2', hue = 'status')
```
