```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from google.colab import files
import io
data = files.upload()
```

```python
df = pd.read_csv(io.StringIO(data['Placement_Data_Full_Class.csv'].decode('utf-8')))
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.suptitle("Graph for different types of Specialisation")
plt.subplot(1,1,1)
plt.gca().set_title('Mkt&HR vs Mkt&Fin')
sns.countplot(x = 'specialisation', palette = 'Set2', data = df)
```

```python
plt.figure(figsize = (15, 5))
plt.suptitle("Analysis Of Placement Status")
plt.subplot(1,1,1)
plt.gca().set_title('specialisation_status')
sns.countplot(x = 'status', hue = 'specialisation', palette = 'Set3', data = df)
```

```python
plt.figure(figsize = (5,5))
sns.histplot(df['specialisation'])
```

```python
plt.figure(figsize = (10,5))
sns.histplot(df['mba_p'])
```

```python
plt.figure(figsize = (10,5))
plt.title("Boxplot of Mba percentage")
sns.boxplot(x=df["mba_p"])
```

```python
plt.figure(figsize = (10,5))
plt.title('Status_Mba_p')
sns.boxplot(x="status", y="mba_p", data=df)
```

```python
plt.figure(figsize = (10, 15))
plt.suptitle("Analysis Of mba_percentage against different variables",fontweight="bold")

plt.subplot(3,2,1)
sns.boxplot(x="gender", y="mba_p", data=df)

plt.subplot(3,2,2)
sns.boxplot(x="hsc_b", y="mba_p", data=df)

plt.subplot(3,2,3)
sns.boxplot(x="ssc_b", y="mba_p", data=df)

plt.subplot(3,2,4)
sns.boxplot(x="hsc_s", y="mba_p", data=df)

plt.subplot(3,2,5)
sns.boxplot(x="degree_t", y="mba_p", data=df)
```

```python
plt.figure(figsize = (10, 15))
plt.suptitle("Analysis Of mba against multiple variables scatterplot",fontweight="bold", fontsize=20)

plt.subplot(2,2,1)
sns.scatterplot(data=df, x="mba_p", y="ssc_p", palette = 'Set1', hue = 'status')

plt.subplot(2,2,2)
sns.scatterplot(data=df, x="mba_p", y="degree_p", palette = 'Set1', hue = 'status')

plt.subplot(2,2,3)
sns.scatterplot(data=df, x="mba_p", y="etest_p", palette = 'Set1', hue = 'status')

plt.subplot(2,2,4)
sns.scatterplot(data=df, x="mba_p", y="hsc_p", palette = 'Set1', hue = 'status')
```
