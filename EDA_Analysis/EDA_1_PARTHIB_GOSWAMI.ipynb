```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('Placement_Data_Full_Class.xls')
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1,1,1)
plt.gca().set_title('count_undergrad_degree')
sns.countplot(x = 'degree_t', palette = 'Set2', data = df)
```

```python
plt.figure(figsize = (15, 5))
plt.suptitle("Analysis Of Placement Status")
plt.subplot(1,1,1)
plt.gca().set_title('degree_type_status')
sns.countplot(x = 'status', hue = 'degree_t', palette = 'Set1', data = df)
```

```python
plt.figure(figsize = (8,5))
sns.histplot(df['degree_p'])
```

```python
plt.title("Degree_test_percentage")
sns.boxplot(x=df["degree_p"])
     
```

```python
plt.figure(figsize = (8, 5))
plt.suptitle("Analysis Of Placement Status: Degree test percentage")
sns.boxplot(x="status", y="degree_p", data=df)
```
