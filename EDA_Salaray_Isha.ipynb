```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from scipy.stats import kendalltau
from scipy.stats import pearsonr
```

```python
dataset=pd.read_csv("Placement_Data_Full_Class[1].csv")
```

```python
dataset.drop(dataset.loc[dataset['status']=='Not Placed'].index,inplace=True)
```

```python
x=dataset["workex"]
y=dataset["salary"]
```

```python
corr,_=kendalltau(x,y)
print(corr);
```

```python
x=dataset["gender"]
y=dataset["salary"]
```

```python
corr,_=kendalltau(x,y)
print(corr);
```

```python
x = dataset['etest_p']
y=dataset["salary"]
```

```python
corr,_=pearsonr(x,y)
print(corr);
```

```python
plt.figure(figsize = (10, 8))
plt.title('Salary vs EmployeeTestPercentage')
sns.scatterplot(x="etest_p", y="salary", data=dataset)
```

```python
plt.figure(figsize = (10, 8))
plt.title('Salary vs WorkExperience')
sns.boxplot(x="workex", y="salary",data=dataset)
```

```python
plt.figure(figsize = (10, 8))
plt.title('Salary vs Gender')
sns.boxplot(x="gender", y="salary",data=dataset)
```

```python
plt.figure(figsize = (10, 8))
plt.title('Salary vs WorkExperience wrt Gender')
sns.scatterplot(x="workex", y="salary",hue = 'gender', data=dataset)
```

```python
plt.figure(figsize = (10, 8))
plt.title('Salary vs EmployeeTestPercentage wrt Gender')
sns.scatterplot(x="etest_p", y="salary",hue = 'gender', data=dataset)
```
