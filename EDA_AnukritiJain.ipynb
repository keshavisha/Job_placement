```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
     

#Read the csv file in DataFrame format
df = pd.read_csv('/Placement_Data_Full_Class.csv')
df.head()
     
```

```python
plt.figure(figsize = (10,5))
#Count of people in various high school boards
plt.subplot(1,3,1)
plt.title('Count_high_school_board')
sns.countplot(x = 'hsc_b', palette = 'Set3', data = df)
#Count of people in different stream 
plt.subplot(1,3,2)
plt.title('Count_high_school_stream')
sns.countplot(x = 'hsc_s', palette = 'Set3', data = df)
```

```python
plt.figure(figsize = (10,5))
sns.histplot(df['hsc_p'])
```

```python
plt.figure(figsize = (10,5))
plt.title("Boxplot high school_percentage")
sns.boxplot(x=df["hsc_p"])
#Boxplot : offers information about a dataâ€™s symmetry, skew, variance, and outliers(the dimond shapes are outliers)
```

```python
plt.figure(figsize = (10, 8))
plt.subplot(2,2,1)
plt.title('Status_high_school_board')
sns.countplot(x = 'status', hue = 'hsc_b', palette = 'Set1', data = df)

plt.subplot(2,2,2)
plt.title('Status_high_school_stream')
sns.countplot(x = 'status', hue = 'hsc_s', palette = 'Set1', data = df)

plt.subplot(2,2,3)
sns.countplot(x = 'gender', hue = 'hsc_b', palette = 'Set1', data = df)

plt.subplot(2,2,4)
sns.countplot(x = 'gender', hue = 'hsc_s', palette = 'Set1', data = df)
```

```python
plt.figure(figsize = (10,5))
plt.title('Status_high_school_board')
sns.boxplot(x="status", y="hsc_p", data=df)
```

```python
plt.figure(figsize = (10, 15))
plt.suptitle("Analysis Of Variable high_school_percentage",fontweight="bold")

plt.subplot(3,2,1)
sns.boxplot(x="gender", y="hsc_p", data=df)

plt.subplot(3,2,2)
sns.boxplot(x="hsc_b", y="hsc_p", data=df)

plt.subplot(3,2,3)
sns.boxplot(x="ssc_b", y="hsc_p", data=df)

plt.subplot(3,2,4)
sns.boxplot(x="hsc_s", y="hsc_p", data=df)

plt.subplot(3,2,5)
sns.boxplot(x="degree_t", y="hsc_p", data=df)
```

```python
plt.figure(figsize = (10, 15))
plt.suptitle("Analysis Of Variable hsc_percentage",fontweight="bold", fontsize=20)

plt.subplot(2,2,1)
sns.scatterplot(data=df, x="hsc_p", y="ssc_p", palette = 'Set2', hue = 'status')

plt.subplot(2,2,2)
sns.scatterplot(data=df, x="hsc_p", y="degree_p", palette = 'Set2', hue = 'status')

plt.subplot(2,2,3)
sns.scatterplot(data=df, x="hsc_p", y="etest_p", palette = 'Set2', hue = 'status')

plt.subplot(2,2,4)
sns.scatterplot(data=df, x="hsc_p", y="mba_p", palette = 'Set2', hue = 'status')
```
